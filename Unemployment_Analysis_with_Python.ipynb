```python
fg=px.histogram(dk,x='States',y='Estimated Unemployment Rate',title='Indian unemployment rate(state-wise)',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
unemployment=dk[["Region",'Estimated Unemployment Rate','States']]
figure=px.sunburst(unemployment,path=["Region","States"],values='Estimated Unemployment Rate',width=700,height=700,title='Indian unemployment rate')
figure.show()
```
